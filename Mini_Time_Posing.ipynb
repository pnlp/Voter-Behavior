# Modeling One 

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
df_2004 = pd.read_csv('./csv_files/df_2004')
df_2008 = pd.read_csv('./csv_files/df_2008')
df_2012 = pd.read_csv('./csv_files/df_2012')
df_2016 = pd.read_csv('./csv_files/df_2016')
```

```python
df_2008['previous_year_total'] = np.log(df_2004['total_vote_ratio'])
df_2008['previous_year_dem'] = np.log(df_2004['dem_vote_ratio'])
df_2008['previous_year_rep'] = np.log(df_2004['rep_vote_ratio'])
df_2008['previous_year_dif'] = df_2004['vote_diff']

df_2008['total_diff'] = df_2008['total_vote_ratio'] - df_2008['previous_year_total']
df_2008['dem_diff'] = df_2008['dem_vote_ratio'] - df_2008['previous_year_dem']
df_2008['rep_diff'] = df_2008['rep_vote_ratio'] - df_2008['previous_year_rep']
df_2008['diff_diff'] = df_2008['vote_diff'] - df_2008['previous_year_dif']


df_2012['previous_year_total'] = np.log(df_2008['total_vote_ratio'])
df_2012['previous_year_dem'] = np.log(df_2008['dem_vote_ratio'])
df_2012['previous_year_rep'] = np.log(df_2008['rep_vote_ratio'])
df_2012['previous_year_dif'] = df_2008['vote_diff']

df_2012['total_diff'] = df_2012['total_vote_ratio'] - df_2012['previous_year_total']
df_2012['dem_diff'] = df_2012['dem_vote_ratio'] - df_2012['previous_year_dem']
df_2012['rep_diff'] = df_2012['rep_vote_ratio'] - df_2012['previous_year_rep']
df_2012['diff_diff'] = df_2012['vote_diff'] - df_2012['previous_year_dif']


df_2016['previous_year_total'] = np.log(df_2012['total_vote_ratio'])
df_2016['previous_year_dem'] = np.log(df_2012['dem_vote_ratio'])
df_2016['previous_year_rep'] = np.log(df_2012['rep_vote_ratio'])
df_2016['previous_year_dif'] = df_2012['vote_diff']

df_2016['total_diff'] = df_2016['total_vote_ratio'] - df_2016['previous_year_total']
df_2016['dem_diff'] = df_2016['dem_vote_ratio'] - df_2016['previous_year_dem']
df_2016['rep_diff'] = df_2016['rep_vote_ratio'] - df_2016['previous_year_rep']
df_2016['diff_diff'] = df_2016['vote_diff'] - df_2016['previous_year_dif']
```

# Linear Model

```python
df_2008.head()
```

```python
def linear_model(df, target):
    
    if target == 'total_vote_ratio' or target == 'total_diff':
        X = df.drop(['total_vote_ratio', 'dem_vote_ratio', 'rep_vote_ratio', 'vote_diff', 
                     'previous_year_dem', 'previous_year_rep', 'previous_year_dif', 
                     'total_diff', 'dem_diff', 'rep_diff', 'diff_diff'], axis=1)
    elif target == 'dem_vote_ratio' or target == 'dem_diff':
        X = df.drop(['total_vote_ratio', 'dem_vote_ratio', 'rep_vote_ratio', 'vote_diff',
                     'previous_year_total', 'previous_year_rep', 'previous_year_dif', 
                     'total_diff', 'dem_diff', 'rep_diff', 'diff_diff'], axis=1)       
    elif target == 'rep_vote_ratio' or target == 'rep_diff':
        X = df.drop(['total_vote_ratio', 'dem_vote_ratio', 'rep_vote_ratio', 'vote_diff',
                     'previous_year_total', 'previous_year_dem', 'previous_year_dif', 
                     'total_diff', 'dem_diff', 'rep_diff', 'diff_diff'], axis=1)
    elif target == 'vote_diff' or target == 'diff_diff':
        X = df.drop(['total_vote_ratio', 'dem_vote_ratio', 'rep_vote_ratio', 'vote_diff',
                     'previous_year_total', 'previous_year_dem', 'previous_year_rep', 
                     'total_diff', 'dem_diff', 'rep_diff', 'diff_diff'], axis=1)
        
    y = df[target]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)

    ss = StandardScaler()
    X_train = ss.fit_transform(X_train)
    X_test = ss.transform(X_test)
    
    lm = LinearRegression()
    lm.fit(X_train, y_train)
    
    print('R^2 on train', lm.score(X_train, y_train).mean())
    print('R^2 on test', cross_val_score(lm, X_test, y_test, cv=5).mean())
    
    predictions = lm.predict(X_test)
    
    print('mean squared error', np.sqrt(mean_squared_error(y_test, predictions)))
    
    if target == 'total_vote_ratio' or target == 'total_diff':
        color = 'rebeccapurple'
    elif target == 'dem_vote_ratio' or target == 'dem_diff':
        color = 'cornflowerblue'
    elif target == 'rep_vote_ratio' or target == 'rep_diff':
        color = 'firebrick'
    else:
        color = 'forestgreen'
    
    plt.scatter(y_test, predictions, alpha=.1, color=color)
    plt.title('Predictions versus Actual');
    plt.show()
    
    residuals = y_test - lm.predict(X_test)
    
    plt.scatter(y_test, residuals, alpha=.1, color=color)
    plt.title('Residuals');
    plt.show()
    
    plt.hist(residuals, bins=25, color=color)
    plt.title("Histogram of Residuals");
    
    columns = []
    for x in X.columns:
        columns.append(x)
    
    feature_importance = pd.DataFrame()
    feature_importance['feature_name'] = columns
    feature_importance['coefficients'] = lm.coef_

    
    fig, ax = plt.subplots()
    fig.set_size_inches(10, 10)
#     plt.xlim(-0.2, 0.2)

    people = feature_importance['feature_name']
    y_pos = np.arange(len(people))
    performance = feature_importance['coefficients']


    ax.barh(y_pos, performance, align='center', color=color)
    ax.set_yticks(y_pos)
    ax.set_yticklabels(people)
    ax.invert_yaxis() 
    plt.title('Feature Importance')
    plt.show()
```

# Random Forest

```python
# possible targets ['total_vote_ratio', 'dem_vote_ratio', 'rep_vote_ratio', 'vote_diff']
    
def forest_model(df, target):
    
    if target == 'total_vote_ratio' or target == 'total_diff':
        X = df.drop(['total_vote_ratio', 'dem_vote_ratio', 'rep_vote_ratio', 'vote_diff', 
                     'previous_year_dem', 'previous_year_rep', 'previous_year_dif', 
                     'total_diff', 'dem_diff', 'rep_diff', 'diff_diff'], axis=1)
    elif target == 'dem_vote_ratio' or target == 'dem_diff':
        X = df.drop(['total_vote_ratio', 'dem_vote_ratio', 'rep_vote_ratio', 'vote_diff',
                     'previous_year_total', 'previous_year_rep', 'previous_year_dif', 
                     'total_diff', 'dem_diff', 'rep_diff', 'diff_diff'], axis=1)       
    elif target == 'rep_vote_ratio' or target == 'rep_diff':
        X = df.drop(['total_vote_ratio', 'dem_vote_ratio', 'rep_vote_ratio', 'vote_diff',
                     'previous_year_total', 'previous_year_dem', 'previous_year_dif', 
                     'total_diff', 'dem_diff', 'rep_diff', 'diff_diff'], axis=1)
    elif target == 'vote_diff' or target == 'diff_diff':
        X = df.drop(['total_vote_ratio', 'dem_vote_ratio', 'rep_vote_ratio', 'vote_diff',
                     'previous_year_total', 'previous_year_dem', 'previous_year_rep', 
                     'total_diff', 'dem_diff', 'rep_diff', 'diff_diff'], axis=1)


    y = df[target]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=27)

    ss = StandardScaler()
    X_train = ss.fit_transform(X_train)
    X_test = ss.transform(X_test)
    
    rf = RandomForestRegressor(n_estimators=50, max_features=10, max_depth=10, min_samples_leaf=10)
    rf.fit(X_train, y_train)
    
    print('R^2 on train', rf.score(X_train, y_train).mean())
    print('R^2 on test', cross_val_score(rf, X_test, y_test, cv=5).mean())
    
    predictions = rf.predict(X_test)
    
    print('mean squared error', np.sqrt(mean_squared_error(y_test, predictions)))
    
    if target == 'total_vote_ratio' or target == 'total_diff':
        color = 'rebeccapurple'
    elif target == 'dem_vote_ratio' or target == 'dem_diff':
        color = 'cornflowerblue'
    elif target == 'rep_vote_ratio' or target == 'rep_diff':
        color = 'firebrick'
    else:
        color = 'forestgreen'
    
    plt.scatter(y_train, rf.predict(X_train), alpha=.1, color=color)
    plt.title('Predictions versus Actual Train');
    plt.show()
    
    plt.scatter(y_test, predictions, alpha=.1, color=color)
    plt.title('Predictions versus Actual Test');
    plt.show()
    
    residuals = y_test - rf.predict(X_test)
    
    plt.scatter(y_test, residuals, alpha=.1, color=color)
    plt.title('Residuals');
    plt.show()
    
    plt.hist(residuals, bins=25, color=color)
    plt.title("Histogram of Residuals");
    
    columns = []
    for x in X.columns:
        columns.append(x)
    
    feature_importance = pd.DataFrame()
    feature_importance['feature_name'] = columns
    feature_importance['features'] = rf.feature_importances_

    
    fig, ax = plt.subplots()
    fig.set_size_inches(10, 10)
#     plt.xlim(-0.2, 0.2)

    people = feature_importance['feature_name']
    y_pos = np.arange(len(people))
    performance = feature_importance['features']


    ax.barh(y_pos, performance, align='center', color=color)
    ax.set_yticks(y_pos)
    ax.set_yticklabels(people)
    ax.invert_yaxis() 
    plt.title('Feature Importance')
    plt.show()
```
